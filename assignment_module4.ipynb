```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data = pd.read_csv("D:\Assignment2\Question_1_output_file.csv")
Dframe=pd.DataFrame(data)#input the file as dataframe
Dframe['Continent']=Dframe['Continent'].replace(np.nan,'NA')#replace np.nan values as NA
newframe1=Dframe[['Estimated HIV Prevalence% - (Ages 15-49)','Continent','1990','2010']]#Frame size decreased
newframe2=newframe1.groupby(['Continent','Estimated HIV Prevalence% - (Ages 15-49)'],as_index=False).mean()#groupby with continent and regions
newframe2
rng = np.random.RandomState(0)
x = rng.randn(50)
y = rng.randn(50)
colors = rng.rand(274)#To get the random color
sizes = 1000 * rng.rand(100)#size of scatter plot
a=plt.figure(figsize=(10,4))#plotting graph of year 1990
plt.scatter(newframe2['Continent'],newframe2['1990'],data=newframe2,c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
b=plt.figure(figsize=(10,4))#plotting graph of year 2010 
plt.scatter(newframe2['Continent'],newframe2['2010'],data=newframe2,c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
```
